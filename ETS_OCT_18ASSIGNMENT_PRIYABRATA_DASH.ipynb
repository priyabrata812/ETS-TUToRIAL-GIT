#***ANALYSIS ON MOLECULAR SOLUBILITY DATA***



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
url="https://raw.githubusercontent.com/GLambard/Molecules_Dataset_Collection/master/latest/ESOL_delaney-processed.csv"
df = pd.read_csv(url, index_col=0)
```

##To see measured solbility variation with molecular weight molecular weight


```python
dot_size=10
sns.regplot(data=df, x='Molecular Weight', y='measured log solubility in mols per litre',color='r',scatter_kws={'color': 'blue','s':dot_size})
plt.show()
```

From the above regression plot, it is completely seen that with increase in molecular weight the solubility of the compound is decreasing for this particular data set. In actual the solubility not only depend upon molecular weight but also on some other factors

##The accuracy of predicted solubility values


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data (replace this with your own DataFrame)
data = df  # Replace 'df' with your DataFrame

# Create a scatter plot
plt.figure(figsize=(8, 6))
sns.regplot(x='measured log solubility in mols per litre', y='ESOL predicted log solubility in mols per litre', data=data)
plt.title('Scatter Plot of Measured vs. Predicted Solubility')
plt.xlabel('Measured Solubility (log M/L)')
plt.ylabel('Predicted Solubility (log M/L)')
plt.show()
```

From the above regression plot of measured vs predicted values of solubility we can clearly see that they are not very much scattered and falls closely to each other
##verifying how close these two with pair wise correlation heat map

```python
import matplotlib.pyplot as plt
import seaborn as sns
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Pairwise Correlation Heatmap')
plt.show()
```

From the heat map now as the pair wise correlation among predicted and measured values of solubility are related or matched with fraction of 0.$96$ i.e $96$% match  refers to a highier value of accuracy for the predicted data
##Seeing the relation between number of Hbond donor and solubility

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Create a 2D density plot
plt.figure(figsize=(8, 6))
sns.kdeplot(
    data=data,
    x='Number of H-Bond Donors',
    y='measured log solubility in mols per litre',
    cmap='Blues',
    fill=True,
)
plt.title('2D Density Plot of H-Bond Donors vs. Measured Solubility')
plt.xlabel('Number of H-Bond Donors')
plt.ylabel('Measured Solubility (log M/L)')
plt.show()
```

From above density plot we can see that with less number of H_bond donor groups the molecules are more soluble, this might be due to absence of intra nuclear H bonding among the molecules itself but in solution they are making H-bond with the solvent

###*These are intentionally done for making some histograms and analysing some random type distribution of data

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select the columns you want to create histograms for
selected_columns = ['Molecular Weight', 'Number of H-Bond Donors', 'Number of Rings']

# Create histograms
for column in selected_columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=df, x=column, kde=True, bins=20)  # Adjust the number of bins as needed
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

1. from molecular weight distribution is it seen that molecules with molecular weight between 100-200 are much in number in the given data.
1. molecules with hydrogen bond donors more than 4 are vwry less from histogram
1. molecules with only one ring is maximum in number and with more than 6 are not present
